```python
import numpy as np
import matplotlib.pyplot as plt

def binary_entropy(p):
    # Use np.where to handle the case when p is 0 or 1
    return np.where(
        p == 0,
        0,
        np.where(
            p == 1,
            0,
            -p * np.log2(p) - (1 - p) * np.log2(1 - p)
        )
    )

def renyi_entripy(p,alpha):
    if alpha ==0: 
        b = np.ones(len(p))
        b[0] = 0
        b[-1] = 0
        return b
    if alpha == 1:
        return binary_entropy(p)
    else:
        p_alpha = p**alpha + (1-p)**alpha
        return np.log2(p_alpha)/(1-alpha)
    
        
    
# Generate values of p from 0 to 1
p_values = np.linspace(0, 1, 1000)
alpha = 0.5

entropy_values = renyi_entripy(p_values,0)
alpha_list = [0,0.1,0.25,0.5,1,2,3,10,100,]

# Plot the binary entropy function
plt.figure(figsize=(10, 6))

for alpha in alpha_list:
    entropy_values = renyi_entripy(p_values,alpha)
    plt.plot(p_values, entropy_values, label=f"alpha = {alpha}")


#plt.plot(p_values, entropy_values, label='Binary Entropy H(p)')
plt.title('Renyi Entropy Function')
plt.xlabel('p')
plt.ylabel('H_alpha(p)')
plt.grid(True)
plt.legend()
plt.show()
```
